```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("../data/raw/expenses.csv")
df.head(10)
```

```python
df.shape
```

```python
df['hour'] = pd.to_datetime(df['createdAt']).dt.hour
df['date'] = pd.to_datetime(df['createdAt']).dt.date
```

```python
df
```

```python
numeric_feature = df[['amount', 'date', 'hour']]
numeric_feature
numeric_feature['date'] = pd.to_datetime(numeric_feature['date']).dt.floor('D')
```

```python
numeric_feature['Day of the week'] = numeric_feature['date'].dt.dayofweek
sorted_by_hour_date = numeric_feature.sort_values(by=['date', 'hour'])
```

```python
sorted_by_hour_date['datetime'] = pd.to_datetime(numeric_feature['date']) + pd.to_timedelta(numeric_feature['hour'], unit='h')
```

```python
sorted_by_hour_date = sorted_by_hour_date.sort_values('datetime')
sorted_by_hour_date['time_gap_minutes'] = sorted_by_hour_date['datetime'].diff().dt.total_seconds() / 3600
sorted_by_hour_date = sorted_by_hour_date.drop(columns=['date', 'datetime'])
```

```python
sorted_by_hour_date = sorted_by_hour_date.fillna(0)
sorted_by_hour_date = sorted_by_hour_date.rename(columns={'Day of the week': 'day_of_week'})
```

```python
sorted_by_hour_date
```

```python
final_numeric_ftr = sorted_by_hour_date
final_numeric_ftr
```

```python
final_numeric_ftr.to_csv('../data/processed/features.csv')
```
